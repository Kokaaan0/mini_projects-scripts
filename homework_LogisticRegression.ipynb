```python
def split_data(data,ratio: list[int,int,int]):
    data_indices = []
    
    for i in range(0,len(data)):
        data_indices.append(i)
    
    train_indices = random.sample(data_indices, k= int(len(data)* ratio[0]))
    for index in train_indices:
        data_indices.remove(index)    
            
    test_indices = random.sample(data_indices, k= int(ratio[1]*len(data)))
    for index in test_indices:
        data_indices.remove(index)
    valid_indices = data_indices    
    train_data = []
    test_data = []
    valid_data = []
    for index in train_indices:
        train_data.append(data[index])
    for index in test_indices:
        test_data.append(data[index])
    for index in valid_indices:
        valid_data.append(data[index])
    train_data = np.array(train_data)
    test_data = np.array(test_data)
    valid_data = np.array(valid_data)
    
    
    x_train = np.array([x[:][:][:-1] for x in train_data])
    y_train = np.array([y[2] for y in train_data])
    
    x_test = np.array([x[:][:][:-1] for x in test_data])
    y_test= np.array([y[2] for y in test_data])
    
    x_valid = np.array([x[:][:][:-1] for x in valid_data])
    y_valid = np.array([y[2] for y in valid_data])
    return x_train,y_train, x_test,y_test, x_valid, y_valid
        
    
```

```python
x_train,y_train, x_test,y_test, x_valid,y_valid = split_data(classification_data,ratio=[0.8,0.1,0.1])
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
    
```

```python
def cost_function(hyphotesis,y):
    cost_f = (-1/len(y)) * np.sum(y-np.log(hyphotesis)+ (1-y)* np.log(1-hyphotesis))
    
```

```python
def hyphotesis(theta1,theta2,x1,x2,b):
    hyphotesis = np.dot(theta1,x1) + np.dot(theta2,x2) + b
    return sigmoid(hyphotesis)
```

```python
def gradient_descent(hyphotesis,y,x,x1,theta,theta1):
    d_theta = x*(hyphotesis-y)
    d_theta1 = x1*(hyphotesis-y)
    d_bias = hyphotesis-y
    return d_theta, d_bias, d_theta1
```

```python
def accuracy(theta,theta1,x,y,b):
    true_predictions = 0
    for i in range(len(y)):
        prediction = hyphotesis(theta,theta1,x[i][0],x[i][1],b)
        if prediction>=0.5:
            prediction = 1
        else:
            prediction = 0
        if prediction == y[i]:
            true_predictions+=1
    accuracy = (true_predictions/ len(y))*100
    return accuracy
            
            
```

```python
def train(X_train, y_train, epochs, alpha,x_valid,y_valid):
    theta1 = 0
    theta = 0
    b=0

    for i in range(0, epochs):

        for j,(x1,x2) in enumerate(X_train):
            
            hyphotesis_ = hyphotesis(theta,theta1,x1,x2,b)
            d_theta,d_bias,d_theta1 = gradient_descent(hyphotesis_,y_train[j],x1,x2,theta,theta1)
            theta-= alpha*d_theta
            theta1-=alpha*d_theta1
            b-= alpha*d_bias
        
        train_acc = accuracy(theta,theta1,x_train,y_train,b)
        valid_acc = accuracy(theta,theta1,x_valid,y_valid,b)
        print('epoch: ',i+1)
        print('train_acc', train_acc)
        print('valid_acc', valid_acc)
        
    return theta,theta1,b


    
```

```python
theta,theta1,b=train(x_train,y_train,epochs=1000,alpha=0.0001,x_valid=x_valid,y_valid=y_valid)
```
