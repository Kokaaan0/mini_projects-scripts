```python
for i in range(len(data)):
    x.append(float(data['x'][i].split()[0]))
    y.append(float(data['x'][i].split()[1]))
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x= x.reshape((100,1))

```

```python
y=y.reshape((100,1))
```

```python
import math
def linear_regression(x,y,w,b,w1):
    x2 = np.cos(x)*np.max(y)
    y_predicted = w*x + b + w1*x2
    
    mse = (np.sum((y-y_predicted)**2))/len(x)
    if mse>0.5:
        w = w-0.0001*np.sum((y_predicted-y)*x)/len(x)
        w1 = w1-0.0001*np.sum((y_predicted-y)*x2)/len(x)
        print('mse')
        print(mse)
        b = b-0.0001*np.sum((y_predicted-y))
        print(w,b,w1)
        return linear_regression(x,y,w,b,w1)
    else:
        return w,b,mse,w1
```

```python
w,b,mse,w1=linear_regression(x,y,1,1,1)
```
